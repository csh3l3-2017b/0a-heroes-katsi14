```python
import pandas as pd
```

```python
hero_attr = pd.read_csv("C:/Users/Katsi/Downloads/Browser/test.csv")
hero_attr
```

```python
define_hero_role = []

for data in range(len(hero_attr)) :
    if hero_attr["movementSpeed"][data] >= 270:
        define_hero_role.append("MARKSMAN")
    elif hero_attr["mana"] [data] >= 500:
        define_hero_role.append("MAGE")
    elif hero_attr["armor"] [data] >= 25:
        define_hero_role.append("TANK")
    elif hero_attr["physicalAttack"] [data] >= 121:
        define_hero_role.append("ASSASSIN")
    elif hero_attr["hp"] [data] >= 2580:
        define_hero_role.append("FIGHTER")
    else: define_hero_role.append("SUPPORT")
```

```python
attr_data = {'name' : hero_attr["name"],
          'role' : define_hero_role}

result = pd.DataFrame(attr_data)
result
```

```python
result.to_csv('prediction.csv',index=False)
```
